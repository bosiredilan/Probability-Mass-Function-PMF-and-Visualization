# Probability Mass Function(PMF)

```python
import collections
import matplotlib.pyplot as plt

# data
list_x = [10,10,10,15,15,15,15,15,15,23,23,23,23,100,100,27,27,27,27]

# Count frequencies
counter = collections.Counter(list_x)
total = len(list_x)

# Keys (unique values) and PMF (probabilities)
keys = list(counter.keys())
pmf = [val/total for val in counter.values()]

# Visualization
plt.bar(keys, pmf, width=5, color="skyblue", edgecolor="black")
plt.xlabel("Outcomes")
plt.ylabel("Probabilities of Outcomes")
plt.title("Probability Mass Function (PMF)")
plt.show()

```

```python
import numpy as np

# Extract values, sorted for neatness
x = sorted(list(counter.elements()))

# Define bins so that each unique value gets its own bar
bins = np.arange(min(x), max(x) + 2) - 0.5  # shift by 0.5 for centered bars

# Plot histogram
plt.hist(x, bins=bins, rwidth=0.6, density=True, color="skyblue", edgecolor="black")

# Set ticks at the unique values
plt.xticks(sorted(counter.keys()))

# Labels and title
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF) Histogram")

plt.show()
```
